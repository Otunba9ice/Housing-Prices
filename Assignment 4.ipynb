```python
#Empty Location

housing_prices[housing_prices.location.isna()]
```

```python
#Return where both pool and gym are null

housing_prices[(housing_prices.Pool.isna()) & (housing_prices.Gym.isna())]
```

```python
#Total number of records and locations

print('Total number of records: ', housing_prices.shape[0])
print('Total number of Locations: ', housing_prices.location.shape[0])
```

```python
#Using notna()

housing_clean_prices = housing_prices[housing_prices.location.notna()]
```

```python
housing_clean_prices.isna().sum()
```

```python
#Changing Yaba to Yaba LGA

housing_clean_prices[housing_clean_prices.location == 'yaba'].location.replace('yaba', 'yaba LGA')
```

```python
housing_clean_prices['location'] = housing_clean_prices.location.replace('yaba','yaba LGA')
```

```python
#To check if the changes has taken effect

housing_clean_prices[housing_clean_prices.location =='yaba LGA'].head(3)
```

```python
#To get the mean

housing_clean_prices.price.mean()
```

```python
#to get the total by locations

housing_clean_prices.groupby('location').price.sum().astype(int).sort_values(ascending = False)
```

```python
#to get the correlation

(housing_clean_prices.corr()).head(4)
```

```python
#to visualize the correlation

sns.heatmap(housing_clean_prices.corr().head(4),annot=True)
```

```python
top_seven = housing_clean_prices.groupby('location').price.sum().sort_values(ascending=False)[:10]

top_seven
```

```python
plt.barh(y=top_seven.index, width=top_seven)

plt.show()
```

```python
plt.pie(top_seven, labels=top_seven.index)

plt.title('Locations by prices')

plt.show()
```

```python
plt.bar(x=top_seven.index, height=top_seven)

plt.title('Location By Prices')

plt.xlabel ('Locations')

plt.ylabel ('prices')

plt.show()
```
